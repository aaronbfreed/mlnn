## 1. Create a 4 dimensional data set with 64 elements and show 2D plots of the data $x_1 \rightarrow y, x_2 \rightarrow y$, etc.

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
l = 64                                                                 #number of elements
n = 4                                                                  #number of variables
m = int(l / n)                                                         #number of datapoints per variable
x = (np.linspace(0,1,l) - np.random.random_sample(l)).reshape(m,n)     #create m x n array of values for x's
y = np.linspace(0,1,m) - np.random.random_sample(m) - 1                #create m x 1 array of values for y
for i in range(0,n):                                                   #plot results
    plt.figure(i)
    plt.scatter(x[:,i],y)
    plt.xlabel("x" + str(i+1))
    plt.ylabel("y")
```

## 2.  Create a model to fit the data. Hint: follow the example from Lesson 3

```python
#note, do not rerun this without first re-running previous block
x = np.insert(x,0,np.ones(m),1)  #add column of 1s to introduce constant term
left = np.linalg.inv(np.dot(x.T, x))
right = np.dot(y.T, x)
np.shape(right)
np.dot(left, right)
```

```python
#checking math
beta = np.linalg.lstsq(x,y,rcond=-1)[0]
beta
```

## 3. Plot the model's prediction in 2D for 2 of the dimensions ($x_1 \rightarrow y_p, x_2 \rightarrow y_p$) along with the original points

```python
pred = np.dot(x,beta)

plt.figure(1)
plt.scatter(x.T[1],y,c="black")
plt.scatter(x.T[1],pred,c="blue")
plt.xlabel("x1")
plt.ylabel("y")

plt.figure(2)
plt.scatter(x.T[2],y,c="black")
plt.scatter(x.T[2],pred,c="red")
plt.xlabel("x2")
plt.ylabel("y")
```

## 4. Read in `mlnn/data/Credit.csv` with Pandas and create a model to predict Credit Rating (`Rating`). Use only the numeric columns in your model, but feel free to experiment which which columns you believe are better predicters of Credit Rating

```python
import pandas as pd
credit = pd.read_csv('../data/Credit.csv')
credit.head()
```

```python
cols = ["Income","Limit"]
X = credit[cols].values
X = np.vstack([X.T, np.ones(len(X))]).T
Y = credit['Rating']
X
```

```python
BETA = np.linalg.lstsq(X,Y,rcond=-1)[0]
BETA
```

### 5. Plot your results (Bonus if you use 3D plots). Show as many of your columns vs. credit rating that you can. 

```python
PRED = np.dot(X,BETA)
plt.figure(1)
plt.scatter(X.T[0],Y,c="black")
plt.scatter(X.T[0],PRED,c="blue")
plt.xlabel(cols[0])
plt.ylabel("Rating")

plt.figure(2)
plt.scatter(X.T[1],Y,c="black")
plt.scatter(X.T[1],PRED,c="red")
plt.xlabel(cols[1])
plt.ylabel("Rating")
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(X.T[0], X.T[1], Y)
ax.scatter3D(X.T[0], X.T[1], PRED, c="red", zdir='z')
ax.set_xlabel(cols[0])
ax.set_ylabel(cols[1])
ax.set_zlabel('Rating')
```
